## SERVIDOR

Podemos tener una computadora que sea nuentro servidor

Hay empresas que alquilan el ordenador ensamblado

Cuando se trabajan datas sensible no lo quieren en la nube por tema de seguridad.

Cunado es una computadora fisica para trabajar, es que si falla la computadora todo el sistema se ve comprometido.

Asi surge la idea de VPS. (Virtual private server)

Amazon EC2 es la tecnologia que se va a utilizar.

De fallar uno de los hardware el VPS no se cae y seguira fucionando.

Si el proyecto escala y comienza tomar mas furza se puede escalar facilmente. Permite agregar RAM y capacidadems de procesamiento sin limites y no tengo que apagar el sistema para agrandar mi servidor privado.

La principal diferencia es el ancho de banda. El datacenter cuenta con un canal de transmision inmenso y es mucho mejor.

Hay muchas empresa que orecen estos servicios.

Servicios de EC2 Amazon, los costos son muy acordes a las prestaciones y accesibles .

Los servicios de Google Cloud  tambien es una opcion importante. Sin embargo, Google ha decidido bloquear los puerto a los servidores de correo para evitar que se utilicen para spam.

En caso de querer organizar sistemas de alertas para informar a los usuarios ante determinados eventos, no podria.
